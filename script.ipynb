# Population Data for US States
Table taken from Wikipedia, "List of states and territories of the United States": https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States

```python
# scrape url
url = "https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# narrowing down the list of tables to the one I'm after
results = soup.find_all("table", class_="wikitable")
results.__len__()
```

```python
# printing the title to find the index I need
for result in results:
    print(result.find("caption").text)
```

```python
# it's the first one
table = results[0]
```

```python
# break it down
rows = table.find_all("tr")
```

```python
# search each row for the elements for state abbreviation and population
# state had \n that needed removing
# population was string with commas, remove them and convert to int
statelist = []
poplist = []

for row in rows:
    try:
        state = row.find("td").text.strip()
        pop = int(row.find("div").text.replace(",",""))
        print(state)
        print(pop)

        if (state and pop):
            statelist.append(state)
            poplist.append(pop)

    except AttributeError as e:
        print(e)
```

```python
# create df and set index
pop_df = pd.DataFrame({
    "state": statelist,
    "population": poplist
})

pop_df.set_index("state", inplace=True)
pop_df.head()
```

# Covid Statistics
Looking at covid data on US Election Day (3 November 2020).

Data taken from Kaggle set "COVID-19 in USA" by SRK, file "us_states_covid19_daily.csv"

https://www.kaggle.com/sudalairajkumar/covid19-in-usa

```python
# read in covid csv
covid_file = "Resources/us_states_covid19_daily.csv"
covid_df = pd.read_csv(covid_file)
covid_df.head()
```

```python
# filter date to required (20201103)
date_df = covid_df.loc[covid_df["date"] == 20201103]
date_df.head()
```

```python
# drop territories (non-states) from df
drop_list = ["AS", "DC", "GU", "MP", "PR", "VI"]

for place in drop_list:
    date_df = date_df.loc[covid_df["state"] != place]

date_df.head()
```

```python
# only want state, positive, and death
covid_cols = ["state", "positive", "death"]
covid_transformed = date_df[covid_cols].copy()

# set index to state
covid_transformed.set_index("state", inplace=True)

covid_transformed.head()
```

# Election Results
Combining county data to reveal the winner of each state.

Data taken from Kaggle set "US Elections Dataset" by Bojan Tunguz, file "1976-2020-president.csv"

https://www.kaggle.com/tunguz/us-elections-dataset

```python
# read in csv
election_file = "Resources/1976-2020-president.csv"
election_df = pd.read_csv(election_file)
election_df.head()
```

```python
# filter for year
recent_election_df = election_df.loc[election_df["year"] == 2020]
recent_election_df.head()
```

```python
# filter to get the candidate with the highest votes per state, then sort back by state for consistency
party_df = recent_election_df.sort_values("candidatevotes", ascending=False).drop_duplicates(["state"])
party_df = party_df.sort_values("state")
party_df.head()
```

```python
# realise I need to drop DC from this set
party_df = party_df.loc[party_df["state_po"] != "DC"]
# party_df.info()
```

```python
# filter df for state_po (renaming to state), candidate, party, candidate votes, and total votes. set index to state
party_cols = ["state_po", "candidate", "party_detailed", "candidatevotes", "totalvotes"]
party_transformed = party_df[party_cols].copy()

party_transformed = party_transformed.rename(columns={"state_po": "state"})

party_transformed.set_index("state", inplace=True)

party_transformed.head()
```
